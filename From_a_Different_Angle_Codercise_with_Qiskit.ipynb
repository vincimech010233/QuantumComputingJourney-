```python
from qiskit import QuantumCircuit, transpile, assemble
from qiskit_aer import AerSimulator
import numpy as np

def apply_rx_pi(state):
    """Apply an RX gate with an angle of π to a particular basis state.

    Args:
        state (int): Either 0 or 1. If 1, initialize the qubit to state |1>
            before applying other operations.

    Returns:
        np.array[complex]: The state of the qubit after the operations.
    """
    qc = QuantumCircuit(1, 1)

    if state == 1:
        qc.x(0)

    qc.rx(np.pi, 0)

    qc.save_statevector()

    qc.measure(0, 0)

    simulator = AerSimulator()

    compiled_qc = transpile(qc, simulator)

    result = simulator.run(compiled_qc).result()

    statevector = result.get_statevector(compiled_qc)
    print(f"Statevector: {statevector}")

    counts = result.get_counts(compiled_qc)
    print(f"Resultados de la medición: {counts}")

    return statevector

print("Estado inicial |0>")
apply_rx_pi(0)

print("Estado inicial |1>")
apply_rx_pi(1)
```

```python
from qiskit import QuantumCircuit, transpile, assemble
from qiskit_aer import AerSimulator
import numpy as np
import matplotlib.pyplot as plt

def apply_rx(theta, state):
    """Apply an RX gate with an angle of theta to a particular basis state.

    Args:
        theta (float): A rotation angle.
        state (int): Either 0 or 1. If 1, initialize the qubit to state |1>
            before applying other operations.

    Returns:
        np.array[complex]: The state of the qubit after the operations.
    """
    qc = QuantumCircuit(1, 1)

    if state == 1:
        qc.x(0)

    qc.rx(theta, 0)

    qc.save_statevector()

    simulator = AerSimulator()

    compiled_qc = transpile(qc, simulator)

    result = simulator.run(compiled_qc).result()

    statevector = result.get_statevector(compiled_qc)
    return statevector

angles = np.linspace(0, 4 * np.pi, 200)

output_states = np.array([apply_rx(t, 0) for t in angles])

def plotter(angles, output_states):
    fig, ax = plt.subplots(2, 1, figsize=(10, 6))
    ax[0].plot(angles, np.real(output_states[:, 0]), label='Re[alpha]')
    ax[0].plot(angles, np.imag(output_states[:, 0]), label='Im[alpha]')
    ax[0].set_ylabel('alpha')
    ax[0].legend()
    ax[1].plot(angles, np.real(output_states[:, 1]), label='Re[beta]')
    ax[1].plot(angles, np.imag(output_states[:, 1]), label='Im[beta]')
    ax[1].set_ylabel('beta')
    ax[1].legend()
    ax[1].set_xlabel('Rotation angle (theta)')
    plt.tight_layout()
    plt.show()

plotter(angles, output_states)
```
